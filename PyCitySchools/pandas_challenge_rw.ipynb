# PyCity Schools Analysis (High Level)
* Below are the two most important trends from my personal analysis on the data. 
    1. The smaller the school size, the better the students perform across all metrics used for this study. If one combines type of school to this equation performance goes up even more. Students at small, charter schools perform the best.
    2. Going against what some may seem as sound education policy, as spending per student goes up, student performance goes down. Spending more money to create better outcomes for students doesn't seem to be the correct route.
* Each subsection of analysis below also has some more detailed findings which contributed to the above high level points
* If further insights were needed from the data, a next best step might be to use visualization along with more robust statistics to evaluate possible trends

```python
# import dependencies
import pandas as pd
```

```python
# files to load
school_file = "resources/schools_complete.csv"
student_file = "resources/students_complete.csv"
```

```python
# read the school and data csv's into pandas and store as dataframes
school_df = pd.read_csv(school_file)
student_df = pd.read_csv(student_file)
```

```python
# show the school dataframe to get an idea of the structure
school_df
```

```python
# show the school dataframe to get an idea of the structure
student_df
```

```python
# create a new df by merging the student and school df's on a left join on school_name 
complete_df = pd.merge(student_df,school_df, how="left", on="school_name")
complete_df
```

# District Summary

```python
#total number of unique schools
unique_schools = school_df["School ID"].count()
unique_schools
```

```python
# total students
tot_students = complete_df["Student ID"].count()
tot_students
```

```python
# total budget
tot_budget = school_df["budget"].sum()
tot_budget
```

```python
# avg. math score
avg_math = complete_df["math_score"].mean()
avg_math
```

```python
# avg. reading score
avg_read = complete_df["reading_score"].mean()
avg_read
```

```python
# % passing math 
pass_math_count = complete_df[(complete_df["math_score"] >= 70)].count()["student_name"]
pass_math_perc = pass_math_count / float(tot_students) * 100
pass_math_perc
```

```python
# % passing reading
pass_read_count = complete_df[(complete_df["reading_score"] >= 70)].count()["student_name"]
pass_read_perc = pass_read_count / float(tot_students) * 100
pass_read_perc
```

```python
# % overall passing (percentage of students passing math and reading)
pass_both_count = complete_df[(complete_df["reading_score"] >= 70) & 
                              (complete_df["math_score"] >= 70)].count()["student_name"]
pass_both_perc = pass_both_count / float(tot_students) * 100
pass_both_perc
```

```python
# create a dataframe for the District Summary
district_summary = pd.DataFrame({"Total Schools":[unique_schools],
                               "Total Students":[tot_students],
                               "Total Budget":[tot_budget],
                               "Average Math Score":[avg_math],
                               "Average Reading Score":[avg_read],
                               "% Passing Math":[pass_math_perc],
                               "% Passing Reading":[pass_read_perc],
                               "% Overall Passing":[pass_both_perc]})

# format dataframe for readabiliity. commas, decimal places, %
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:.2f}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:.2f}".format)
district_summary["% Passing Math"] = district_summary["% Passing Math"].map("{:.2f}%".format)
district_summary["% Passing Reading"] = district_summary["% Passing Reading"].map("{:.2f}%".format)
district_summary["% Overall Passing"] = district_summary["% Overall Passing"].map("{:.2f}%".format)

# print summary dataframe
district_summary
```

### Analysis
* Reading scores district wide are higher than math scores and while the average of all reading scores is ~3 points higher than the average of all math scores, the difference between those passing reading vs those passing math is over 10 points. This leads me to think there may be some outliers within the data and thus suggest further statistical investigation (i.e. median, standard deviation, quartiles)
* % Overall Passing results leads to a possible conclusion that students are more likely to test better at either math or reading than testing well on both subjects. Well-roundedness may be an issue to explore more, district wide. 

## School Summary


```python
# code provided from starter file
school_types = school_df.set_index(["school_name"])["type"]
school_types
```

```python
# total students per school
students_per_school = complete_df.groupby("school_name").count()["Student ID"]
students_per_school
```

```python
# total budget per school
budget_per_school = complete_df.groupby("school_name").mean()["budget"]
```

```python
# budget per student for each school
per_capita_budget = budget_per_school/students_per_school
per_capita_budget
```

```python
# average math score per school
avg_math_school = complete_df.groupby("school_name").mean()["math_score"]
avg_math_school 
```

```python
# average reading score per school
avg_read_school = complete_df.groupby("school_name").mean()["reading_score"]
avg_read_school
```

```python
# % passing math per school

# create dataframe of students with only math scores > 70
pass_math_school = complete_df.loc[(complete_df["math_score"] >= 70)]

# groupby school name and student name count then divide by total student
# count and multiply by 100 to create percentage
pass_math_group = pass_math_school.groupby("school_name").count()["student_name"]/students_per_school*100
pass_math_group

```

```python
# % passing reading per school

# create dataframe of students with only reading scores > 70
pass_read_school = complete_df.loc[(complete_df["reading_score"] >= 70)]

# groupby school name and student name count then divide by total student
# count and multiply by 100 to create percentage
pass_read_group = pass_read_school.groupby("school_name").count()["student_name"]/students_per_school*100
pass_read_group
```

```python
# % overall passing per school

# create dataframe of students with only math scores > 70 and reading scores > 70
pass_overall_school = complete_df.loc[(complete_df["reading_score"] >= 70) &
                                  (complete_df["math_score"] >= 70)]

# groupby school name and student name count then divide by total student
# count and multiply by 100 to create percentage
pass_overall_group = pass_overall_school.groupby("school_name").count()["student_name"]/students_per_school*100
pass_overall_group
```

```python
# create a dataframe for the School Summary

per_school_summary = pd.DataFrame({"School Type":school_types,
                               "Total Students":students_per_school,
                               "Total School Budget":budget_per_school,
                              "Per Student Budget":per_capita_budget,
                              "Average Math Score":avg_math_school,
                              "Average Reading Score":avg_read_school,
                              "% Passing Math":pass_math_group,
                              "% Passing Reading":pass_read_group,
                              "% Overall Passing":pass_overall_group})


# format dataframe for readabiliity. commas, decimal places, %
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

# get rid of index name for readability purposes
per_school_summary.index.name = None

# print summary dataframe
per_school_summary
```

### Analysis
* Total School Budgets can vary widely (millions) as can the size of student populations. Therefore these metrics should not be looked at in a vacuum and metrics like Per Student Budget should be given more weight when analyzing spending affects on academic outcomes.
* As in the District Summary, the average metrics don't vary too much from subject to subject and school to school, approximately in the range of 76 - 83. On the other hand, the % passing metrics vary widely, especially in math and overall passing. This once again begs the question of whether average is a good indicator of academic performance for our schools. More statistics are needed to test this assumption.

## Scores by School Spending

```python
# create bins and labels
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# created a new dataframe (same as school_summary) to work around the 
# formatting issues from just copying the school_summary dataframe
school_spending_df = pd.DataFrame({"School Type":school_types,
                               "Total Students":students_per_school,
                               "Total School Budget":budget_per_school,
                              "Per Student Budget":per_capita_budget,
                              "Average Math Score":avg_math_school,
                              "Average Reading Score":avg_read_school,
                              "% Passing Math":pass_math_group,
                              "% Passing Reading":pass_read_group,
                              "% Overall Passing":pass_overall_group})
```

```python
# check data types
school_spending_df.dtypes
```

```python
# use pd.cut on school_spending_df["Per Student Budget"] in order to created binned ranges of spending and new column
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Per Student Budget"], 
                                                             bins=spending_bins, labels=labels, include_lowest=True)
# display results
school_spending_df
```

```python
# create mean scores by metric for each spending range
spend_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]
spend_read_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
spend_pass_math = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]
spend_pass_read = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
spend_pass_overall = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]
```

```python
# create dataframe with mean scores by metric for each spending range
spending_summary = pd.DataFrame({"Average Math Score":spend_math_scores,
                                "Average Reading Score":spend_read_scores,
                                "% Passing Math":spend_pass_math,
                                "% Passing Reading":spend_pass_read,
                                 "% Overall Passing":spend_pass_overall})

# display results
spending_summary
```

### Analysis
* There is a negative linear correlation between spending per student and student performance across all metrics. As spending per student increases student performance goes down. This trend seems to go against what many might think as common sense education policy; spend more money per student and performance will go up.

## Scores by School Size

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# use pd.cut on school_summary["School Size"] in order to created binned ranges of size and new column
per_school_summary["School Size"] = pd.cut(per_school_summary["Total Students"], 
                                                             bins=size_bins, labels=labels, include_lowest=True)
# display results
per_school_summary
```

```python
# create mean scores by metric for each size range
size_math_scores = per_school_summary.groupby(["School Size"]).mean()["Average Math Score"]
size_read_scores = per_school_summary.groupby(["School Size"]).mean()["Average Reading Score"]
size_pass_math = per_school_summary.groupby(["School Size"]).mean()["% Passing Math"]
size_pass_read = per_school_summary.groupby(["School Size"]).mean()["% Passing Reading"]
size_pass_overall = per_school_summary.groupby(["School Size"]).mean()["% Overall Passing"]
```

```python
# create dataframe with mean scores by metric for each size range
size_summary = pd.DataFrame({"Average Math Score":size_math_scores,
                                "Average Reading Score":size_read_scores,
                                "% Passing Math":size_pass_math,
                                "% Passing Reading":size_pass_read,
                                 "% Overall Passing":size_pass_overall})

# display results
size_summary
```

### Analysis
* Across all performance metrics, except Average Reading Score, small(<1000) and medium(1000-2000) sized schools perform much better than large(2000-5000) sized schools. Large schools still don't perform as well in Average Reading Score, but the difference is ~2 points compared to differences as big as ~30 points (% Overall Passing).
